3. Describe three different data types.
Solution:
Data form or Data Type are classification of data items. It represent the value which tells what kind of operations can be performed in those data items.
Standard Build-in Data type in python:

1. Numeric type
2.Sequence
3.Boolean
4.Set
5. Dictionary

1. Numeric type- As the name defines, it contain numerical values such as integers, float numbers i.e. decimal number and complex numbers represented by
int class- containing all the negative and positive numbers,
float- containing floating or decimal numbers 
complex class- containing complex number, both real and imaginary parts i.e. 2+ 6j

2.Sequence type- Sequence is ordered collections of similar or different types of values. Type of sequence are string, tuple, list etc
String- A string is a collection of one or more characters put in a single quote, double-quote or triple quote.
List : List are just like arrays, declared in othet languages which is a ordered collection of data. It is very flexible as the items in a list do not need to be of the same type. 
Tuple : Tuple is also an ordered collection of Python objects. The only difference between tuple and list is that tuple are immutable i.e. tuples cannnot be modified after it is created. It is represented by tuple class.

3.Boolean: this data type contain two build in values which are True and False. True and False with capital ‘T’ and ‘F’ are valid booleans otherwise python will throw an error. Bollean operation is performed usinh Bool.

4.Set- Set is an unordered collection of data type that is iterable, mutable and has no duplicate elements. The order of elements in a set is undefined though it may consist of various elements.The major advantage of using a set, as opposed to a list, is that it has a highly optimized method for checking whether a specific element is contained in the set.Sets can be created by using the built-in set() function with an iterable object or a sequence by placing the sequence inside curly braces, separated by ‘comma’.
