17. How do you distinguish between copy.copy() and copy.deepcopy()?
Solution:A shallow copy i.e. copy.copy() constructs a new compound object and then (to the extent possible) inserts references into it to the objects found in the original. A deep copy i.e. copy.deepcopy() constructs a new compound object and then, recursively, inserts copies into it of the objects found in the original.
